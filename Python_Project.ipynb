```python
df['Amount']=df['Amount'].astype('int')
#change the data type of amount into integer
```

```python
df.columns
```

```python
sns.countplot(x='Gender',data=df)
```

```python
ax=sns.countplot(x='Gender',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
#by this graph we can see female have done more purchase
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sns.barplot(x='Gender',y='Amount',data=sales_gen);
#in shopping womens spent much money
```

```python
ax=sns.countplot(x='Age Group',data=df,hue='Gender')
for bars in ax.containers:
     ax.bar_label(bars)
```

```python
sales_df=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sns.barplot(x='Age Group',y='Amount',data=sales_df);
#from this graph we can see most of the buyers are of age group between 26-35 yrs female
```

```python
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
```

```python
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales_state,x='State',y='Orders');
#rc is used for runtime configuration
```

```python
sales_state=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
```

```python
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales_state,x='Occupation',y='Amount');
```

```python
sns.set(rc={'figure.figsize':(25,5)})
ax=sns.countplot(data=df,x='Product_Category')
for bars in ax.containers:
    ax.bar_label(bars)
#most sold product food
```

```python
#Conclusion // married women age group 26-35 yrs from up,maharashtra and karnataka working in IT,healthcare and Aviation are more likely to buy products from food ,clothing and electronics category.
```
